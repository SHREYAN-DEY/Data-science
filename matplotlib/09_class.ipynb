### Matplotlib with Pandas

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/pokemon.csv")
print(df)
```

```python
print(df["Type1"].value_counts())
```

```python
type_count = df["Type1"].value_counts(ascending=True)

plt.barh(type_count.index, type_count.values, color="red",
                                              edgecolor="black")
plt.title("No of pokemon by primary type")
plt.xlabel("Count")
plt.ylabel("Type")

plt.tight_layout()
plt.show()
```
