### Titles, labels and ticks : 

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Creation of basic plot
x = np.array([2023, 2024, 2025, 2026])
y1 = np.array([15, 25, 30, 20])
y2 = np.array([17, 23, 38, 5])

# For adding and modifying titles, lables and ticks
plt.title("Class size", fontsize=20,
                        family="Arial",
                        fontweight="bold",
                        color="#2d4cfc")

plt.xlabel("Year", fontsize=15,
                   family="Arial",
                   fontweight="bold",
                   color="#2d4cfc")

plt.ylabel("Students",fontsize=15,
                      family="Arial",
                      fontweight="bold",
                      color="#2d4cfc")

lineStyle = dict(marker=".", 
               markersize=10,
               markerfacecolor="red",
               markeredgecolor="red",
               linestyle="solid",
               linewidth="3",
               color="black")

plt.tick_params(axis="both", colors="brown")

# Adjusting padding for labels
plt.subplots_adjust(bottom=0.07, left=0.07)

# plotting the graph
plt.plot(x, y1, **lineStyle)
plt.plot(x, y2, **lineStyle)

# modifying the values in the x axis (ticks)
plt.xticks(x)

# To show the output we have to call the show() function
plt.show()
```
