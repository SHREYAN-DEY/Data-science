### Bar chart :

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
catagories = np.array(["Grain", "Fruit", "Vegetables", "Protein", "Dairy", "Sweets"])
values = np.array([4, 3, 2, 5, 3, 1])

plt.bar(catagories, values)

plt.title("Daily consumption bar chart")
plt.xlabel("Food types")
plt.ylabel("Quantities")

plt.show()
```
