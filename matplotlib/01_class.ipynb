```python
import matplotlib as mplt
import matplotlib.pyplot as plt
import numpy as np
print("Matplotlib version : ",mplt.__version__)
```

```python
# Creation of basic plot
x = np.array([2023, 2024, 2025, 2026])
y1 = np.array([15, 25, 30, 20])
y2 = np.array([17, 23, 38, 5])

lineStyle = dict(marker=".", 
               markersize=10,
               markerfacecolor="red",
               markeredgecolor="red",
               linestyle="solid",
               linewidth="3",
               color="black")

plt.plot(x, y1, **lineStyle)

plt.plot(x, y2, **lineStyle)

# To show the output we have to call the show() function
plt.show()
```
